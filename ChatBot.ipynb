```python
from transformers import pipeline
```

```python
chat = [
    {"role": "system", "content": "I am an helpful Ai assistant."},
    {"role": "user", "content": "Hey, can you help me with my homework?"}
]
```

```python
pipe = pipeline("text-generation", model="Qwen/Qwen2-0.5B-Instruct")
```

```python
resp = pipe(chat, max_new_tokens=512)
resp
```

```python
def chat_bot_func(Prompt):
  chat.append({'role': 'user', 'content': Prompt})
  resp = pipe(chat, max_new_tokens=512)
  chat.append({'role': 'system', 'content': resp[0]['generated_text'][-1]['content']})
  return resp[0]['generated_text'][-1]['content']
```

```python
import gradio as gr
```

```python
demo = gr.Interface(fn=chat_bot_func, inputs="text", outputs="text", title="A simple ChatBot", description="Let AI model conversate with you. It remembers what you say.....")
```

```python
demo.launch()
```

```python
fb = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
response = fb("The sky appears blue because of a phenomenon called Rayleigh scattering. When sunlight enters Earth’s atmosphere, it collides with molecules and small particles in the air. Sunlight is made up of different colors, each with its own wavelength. Blue light has shorter, smaller waves compared to other colors. These shorter waves are scattered in all directions by the gases and particles in the atmosphere. Because blue light is scattered more than other colors, the sky looks blue to our eyes during the day. However, during sunrise and sunset, the sky can appear red or orange because the light has to pass through more of the atmosphere, scattering the shorter blue wavelengths and allowing the longer red wavelengths to reach our eyes", max_length=30, min_length=20)
response
```

```python
def summarize_func(text):
  size = len(text.split())
  response = fb(text, max_length=int(size*0.3), min_length=int(size*0.2))
  return response[0]['summary_text']
```

```python
summarizer = gr.Interface(fn=summarize_func, inputs="text", outputs="text", title="A simple Summarizer", description="Let AI model summarize your text")
```

```python
summarizer.launch()
```
