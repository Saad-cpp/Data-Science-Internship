```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings, SimpleDirectoryReader, VectorStoreIndex
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.postprocessor import SimilarityPostprocessor
```

```python
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
```

```python
Settings.llm = None
Settings.chunk_size = 256
Settings.chunk_overlap = 25
```

```python
document = SimpleDirectoryReader(input_files=["cats.txt"]).load_data()
```

```python
index = VectorStoreIndex.from_documents(document)
```

```python
retriever = VectorIndexRetriever(index=index)
```

```python
query_engine = RetrieverQueryEngine(
    retriever=retriever,
    node_postprocessors=[
        SimilarityPostprocessor(similarity_cutoff=0.5)
    ]
)
```

```python
query = "How many types of cats are there?"
response = query_engine.query(query)
print(response)
```

```python
response.source_nodes[0].text
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
model_name = "TheBloke/Mistral-7B-Instruct-v0.2-GPTQ"
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", trust_remote_code=False, revision="main")
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
```

```python
context = "Context:\n"
for node in response.source_nodes:
    context += node.text + "\n"
print(context)
```

```python
prompt_template = lambda context, query: f"""[INST] You are a CatGPT that teaches general knowledge about cats. Make your answers detailed and in line with given context. DO NOT SAY ANYTHING OUTSIDE OF THE CONTEXT.
{context}
now answer the given question:
Question: {query} [/INST]"""
```

```python
question = "How many types of cats are there?"
prompt = prompt_template(context, question)
```

```python
input_ids = tokenizer.encode(prompt, return_tensors="pt").to("cuda")
```

```python
output_ids = model.generate(input_ids.to("cuda"), max_new_tokens=256)
```

```python
response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print(response)
```

```python
print(response.split("[/INST]")[1])
```

```python
question = "How many types of wild cats are there?"
prompt = prompt_template(context, question)
input_ids = tokenizer.encode(prompt, return_tensors="pt").to("cuda")
output_ids = model.generate(input_ids.to("cuda"), max_new_tokens=256)
response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
```

```python
print(response.split("[/INST]")[1])
```
